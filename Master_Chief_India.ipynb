```python
# Preprocess the dataset
ingredients = set()
recipe_names = set()
for recipe in recipes:
    recipe_name, ingredient_list = recipe.split(":")
    recipe_names.add(recipe_name.strip())
    recipe_ingredients = [ingredient.strip() for ingredient in ingredient_list.split()]
    ingredients.update(recipe_ingredients)

ingredient_vocab = {ingredient: idx + 1 for idx, ingredient in enumerate(sorted(ingredients))}
reverse_vocab = {idx + 1: ingredient for ingredient, idx in ingredient_vocab.items()}
recipe_vocab = {recipe_name: idx for idx, recipe_name in enumerate(sorted(recipe_names))}
reverse_recipe_vocab = {idx: recipe_name for recipe_name, idx in recipe_vocab.items()}
num_ingredients = len(ingredient_vocab)
num_recipes = len(recipe_vocab)

# Convert recipes to input and output sequences
input_sequences = []
output_sequences = []
for recipe in recipes:
    recipe_name, ingredient_list = recipe.split(":")
    recipe_ingredients = [ingredient.strip() for ingredient in ingredient_list.split()]
    input_sequence = [ingredient_vocab[ingredient] for ingredient in recipe_ingredients]
    output_sequence = [recipe_vocab[recipe_name.strip()]]
    input_sequences.append(input_sequence)
    output_sequences.append(output_sequence)

# Pad sequences to have the same length
max_sequence_length = max(len(sequence) for sequence in input_sequences)
input_sequences = np.array([sequence + [0] * (max_sequence_length - len(sequence)) for sequence in input_sequences])
output_sequences = np.array(output_sequences)

# Define the model architecture
model = Sequential()
model.add(Embedding(num_ingredients+1, 100, input_length=max_sequence_length))
model.add(LSTM(100))
model.add(Dense(num_recipes, activation='softmax'))

# Compile and train the model
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam')
model.fit(input_sequences, output_sequences, epochs=1000, verbose=1)

model.save_weights('/content/sample_data/model.h5')

# Generate recipe suggestions
def suggest_recipe(ingredients):
    input_sequence = [ingredient_vocab[ingredient] for ingredient in ingredients]
    input_sequence = np.array(input_sequence)
    input_sequence = np.pad(input_sequence, (0, max_sequence_length - len(input_sequence)), 'constant')
    predicted_probabilities = model.predict(np.array([input_sequence]))[0]
    predicted_class = np.argmax(predicted_probabilities)
    recipe_name = reverse_recipe_vocab[predicted_class]
    return recipe_name
```

```python
# Test the model
test_ingredients = ["Milk-powder", "plain-flour", "baking-powder", "ghee","rose-water"]
suggested_recipe = suggest_recipe(test_ingredients)
print(f"Suggested Recipe: {suggested_recipe}")
```
